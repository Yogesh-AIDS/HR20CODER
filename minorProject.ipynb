```python
batsman = pd.read_csv("/content/sharma-kohli.csv")
batsman
plt.plot(batsman['index'],batsman['V Kohli'])
plt.show
```

```python
#ploting multiple graph using the data
```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])
plt.show()
```

```python
# labels title

```

```python
plt.plot(batsman['index'],batsman['V Kohli'])
plt.plot(batsman['index'],batsman['RG Sharma'])

plt.title('V Kholi vs RG Sharma')
plt.xlabel('year')
plt.ylabel('Runs')

plt.show()
```

```python
# colour
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='red')
plt.plot(batsman['index'],batsman['RG Sharma'],color ='blue')

plt.title('V Kholi vs RG Sharma')
plt.xlabel('year')
plt.ylabel('Runs')

plt.show()
```

```python
# we can add the lineeidth and the linestyle for checking the perticular thing
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='solid',linewidth=3)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2)

plt.title('V Kholi vs RG Sharma')
plt.xlabel('year')
plt.ylabel('Runs')

plt.show()
```

```python
# give the markersize and thwe marker at the corner
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='solid',linewidth=3 ,marker='d',markersize=10)
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='o')

plt.title('V Kholi vs RG Sharma')
plt.xlabel('year')
plt.ylabel('Runs')

plt.show()
```

```python
# adding the label to assign the name using = plt.legend(loc='upper left')
```

```python
plt.plot(batsman['index'],batsman['V Kohli'],color='blue',linestyle='solid',linewidth=3 ,marker='d',markersize=10, label='Virat')
plt.plot(batsman['index'],batsman['RG Sharma'],color='red',linestyle='dashdot',linewidth=2,marker='o',label='Rohit' )

plt.title('V Kholi vs RG Sharma')
plt.xlabel('year')
plt.ylabel('Runs')
plt.legend(loc='upper left')
plt.show()
```

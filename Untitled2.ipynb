```python
df = pd.read_csv('/content/batter.csv')
df = df.head(50)
df
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='batter', y='runs', data=df)
plt.xticks(rotation=90)
plt.title('Runs per Batter')
plt.xlabel('Batter')
plt.ylabel('Runs')
plt.tight_layout()
plt.show()
```
